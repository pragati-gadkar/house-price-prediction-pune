```python
a = data1.drop('Price', axis=1)
b = data1['Price']
```

```python
# Split the training set into 
from sklearn.model_selection import train_test_split
# training and validation set
a_train, a_valid, b_train, b_valid = train_test_split( a, b, train_size=0.95, test_size=0.05, random_state=0)
```

```python
from sklearn.metrics import mean_absolute_percentage_error, mean_absolute_error, mean_squared_error, r2_score

def evaluate_model(predictions, actual):
    mape = mean_absolute_percentage_error(actual, predictions)
    mae = mean_absolute_error(actual, predictions)
    mse = mean_squared_error(actual, predictions)
    r2 = r2_score(actual, predictions)
    return mape, mae, mse, r2
```

```python
# Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor

model_DT = DecisionTreeRegressor(random_state=0)
model_DT.fit(a_train, b_train)
b_pred_DT = model_DT.predict(a_valid)

mape_DT, mae_DT, mse_DT, r2_DT = evaluate_model(b_pred_DT, b_valid)

print("Decision Tree Regressor:")
print(f"MAPE: {mape_DT}")
print(f"MAE: {mae_DT}")
print(f"MSE: {mse_DT}")
print(f"R-squared: {r2_DT}")

# Assuming you have a threshold for categorization (e.g., 0.1)
threshold = 0.075

# Categorize predictions based on the threshold
correct_predictions = sum(abs(b_pred_DT - b_valid) / b_valid <= threshold)
incorrect_predictions = len(b_valid) - correct_predictions

# Data for the pie chart
labels = ['Correct Predictions', 'Incorrect Predictions']
sizes = [correct_predictions, incorrect_predictions]

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.

plt.title('Decision Tree Regressor: Correct vs. Incorrect Predictions')
plt.show()

categories = ['Correct Predictions', 'Incorrect Predictions']
counts = [correct_predictions, incorrect_predictions]

plt.bar(categories, counts, color=['green', 'red'])
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Decision Tree Regressor: Correct vs. Incorrect Predictions')
plt.xticks(rotation=0)  # To prevent the category labels from being rotated

plt.show()
```
